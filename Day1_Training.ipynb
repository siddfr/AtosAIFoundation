```
num1 = int(input("enter first number"))
num2 = int(input("enter second number"))
operation = input("enter operation")
if operation == "add":
  answer = num1 + num2
elif operation == "sub":
  answer = num1 - num2
elif operation == "mul":
  answer = num1 * num2
elif operation == "div":
  # Basic division, consider adding error handling for division by zero
  if num2 != 0:
    answer = num1 / num2
  else:
    print("Error: Division by zero!")
    answer = 0 # Or handle as per requirement
else:
  print("Unknown operation")
  answer = 0
print(answer)
```
